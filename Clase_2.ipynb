# Introducción a métodos

Se les llama métodos a las funciones que pertenecen a una clase, por ahora no necesitamos ahondar en la definición de clase, pero resumidamente, es un conjunto de funciones y variables específicas para un tipo de cosa. Por ejemplo, la clase "carro" podría tener la variable "placa" y el método "manejar"

### Sintaxis
La diferencia elemental entre un método y una función es que escriben de forma distinta.

La función se escribe de la forma: función(argumento).


Un método se escribiría así: variable.función(argumento), es decir, se usa una variable que pertenezca a una clase, y se invoca mediante el uso de un punto. También se podría escribir de la manera clase.función(variable).

image.png

```python
#Los tipos de variables que hemos visto (int, float, str) son clases, y para cada uno hay funciones específicas (métodos)
#Veremos específicamente varios métodos para las cadenas

cadena_9 = 'hola, mundo'

#Para recordar, ejemplo de una función:
print(cadena_9)

#El primer método que vamos a ver: capitalize
print(cadena_9.capitalize())
```

Como se puede observar, el método capitalize() coloca en mayúsculas la primera letra de la cadena, y empezó invocándose la variable, para luego conectarla con un punto al método

```python
#No se necesita una variable para usar los métodos, siempre y cuando se use el valor de la clase correcta

print('colombia con mayúscula'.capitalize())
```

```python
#La otra manera de usar métodos es invocando la clase de la variable, y luego llamar a esta en el argumento
#recomendaría usar la primera opción

print(str.capitalize('colombia con mayúscula'))
```
