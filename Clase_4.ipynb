# Diccionarios y tuplas

Aparte de las listas, existen otras formas de guardar información



*   Diccionarios: guardan los valores con claves y se pueden editar
*   Tuplas: son como listas que no se pueden editar, ideal para valores fijos que nunca van a cambiar durante el código



```python
diccionario = {}
diccionario["edad"] = 16
diccionario["color favorito"] = "azul"
print(diccionario, diccionario["edad"])
```

```python
tupla = (1, 2, 3, 4)
print(tupla)
```

# Listas por comprensión

se pueden resumir los ciclos iterativos en una lista

```python
#forma explícita:

lista = []
for i in range(0, 10, 2):
  lista.append(i)

#forma rápida
lista_rapida = [i for i in range(0, 10, 2)]

print(lista == lista_rapida)
```

También pueden agregar condicionales


```python
pares = []

for i in range(100):
  if i%2 == 0:
    pares.append(i)

pares_rapido = [i for i in range(100) if i%2 == 0]

print(pares == pares_rapido)
```

```python
cuadrados_de_impares = []

for i in range(30):
  if i%2 != 0:
    cuadrados_de_impares.append(i**2)

cuadrados_de_impares_rapido = [i**2 for i in range(30) if i%2 != 0]

print(cuadrados_de_impares == cuadrados_de_impares_rapido)
```

# Funciones

Ya han usado funciones, es hora de ver cómo se crean

Sus partes son:



1.   La declaración de la función (palabra clave "def" + nombre)
2.   Argumento de la función (que valores recibirá)
3.   Lógica del código
4.   Transformación del argumento o devolución de valor



```python
def suma(a, b): #paso 1 y 2
  return a+b #paso 3 y 4

print(suma(4, 3))
```

```python
listado = [2, 3, 4, 3, 4, 3, 4, 3, 3, 4]

def promedio(valores):
  promedio = sum(valores)/len(valores)
  return promedio

print(promedio(listado))
```

```python
def varianza(valores):
  avg = promedio(valores)
  return sum([(valor-avg)**2 for valor in valores])/len(valores)

print(varianza(listado))
```

```python
def saludo(nombre):
  return f"¡Hola {nombre}!"

print(saludo("Juan"))
```
