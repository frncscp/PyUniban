# Condicionales

Muchas veces nos encontraremos en situaciones donde querremos que un código se ejecute "solo si" una condición se cumple. Por ejemplo, solo si eres mayor de edad, deberías poder votar.

```python
edad = int(input('Inserte su edad: '))

if edad >= 18:
  print('Usted es un adulto, puede votar.')
else:
  print('Los menores de edad no pueden votar.')
```

Revisemos el código de arriba, está compuesto por:

1. Una variable llamada "edad", que tiene asignada una entrada de teclado solo de números enteros, donde se le pregunta al usuario por su edad.

2. if edad >= 18, lo que quiere decir es: "si la edad es 18 o más", if es siempre la primera condición que se quiere hacer. Como se puede observar, el print después del condicional va anidado, esto quiere decir que hay que presionar la tecla "tab" y dejar un espacio antes de escribir, por cada condicional que se use, un tab que se debe pulsar.

3. else, significa "de lo contrario", en este caso, lo único que puede pasar si no se es mayor de edad, es que se sea menor de edad.

```python
sueldo = int(input("Inserte su sueldo: "))

if sueldo > 3000:
  print("Esta persona debe abonar impuestos")
```

en este caso no se usa else, ya que en el diagrama no hay una respuesta por si el sueldo es menor a 3000, pero podemos aprovechar este ejemplo para introducir a un nuevo condicional.

```python
sueldo = int(input("Inserte su sueldo: "))

if sueldo > 3000:
  print("Esta persona debe abonar impuestos del 60%")

elif sueldo > 2000:
  print("Esta persona debe abonar impuestos del 40%")

elif sueldo < 1000:
  print("Esta persona no debe abonar impuestos")
```

Aquí se introduce un nuevo condicional, elif, su función es poder repetirse, lo que no puede hacer el if, ya que solo puede utilizarse la primera vez.

```python
edad = int(input("Escribe tu edad: ")) #entrada de teclado numérica donde se pide la edad

if edad >= 18: #si eres mayor de 18 años (o tienes 18), no deberías estar en el colegio
  print('Probablemente no estudias en el colegio')

else: #de lo contrario (si tu edad es menor a 18 años)
  #se hace una anidación porque se está iniciando un nuevo condicional (if)

  if edad >= 5: #si tienes más de 5 años (edad mínima para entrar a primero)
    print('Eres un estudiante de colegio') 
    #sabemos que debe ser estudiante de colegio porque es menor de 18 y también mayor de 5 años (o los tiene)
    
    if edad > 11: 
      #Ya sabemos que tiene más de 5 (o los tiene) y tiene menos de 18, pero si es mayor a 11 (12-17) sabemos que está en bachillerato
      print('Debes estar en bachillerato')

    else: #Si tiene más de 5 años (o los tiene) y menos de 18, y además es menor de 11 años (5-11) sabemos que está en primaria
      print('Debes estar en primaria')

  else: #Si no es mayor de 5 años (o los tiene), no pudo haber entrado al colegio
    print('No has entrado al colegio')
```

# Listas

Antes de empezar a ver bucles, es importante ver este tipo de almacenamiento de datos, ya que los van a usar mucho.

Ocurre que no siempre sabemos cuantos valores necesitamos, o queremos guardar varios datos en una sola variable, pues esto es posible, gracias a las listas, tuplas, sets y diccionarios, para esta clase solo veremos las listas.

Las listas son un conjunto de valores ordenado, esto quiere decir que tienen un orden y se pueden indexar por posición (como las cadenas).

```python
#Una cadena es muy similar a una lista
#Los dos se pueden indexar por posición
#Esto quiere decir que podemos acceder al primer elemento por ejemplo

cadena = 'Hola, mundo!'
lista = ['Hola', 3, True, 3+1j]

print(cadena[0], lista[-1])
```

***Reglas básicas para indexar***

1. Para indexar se debe colocar el nombre de la variable junto a unos llaves [ ] que lleven dentro el valor de posición.

2. Solo las cadenas y almacenadoras de varios valores (como las listas) se pueden indexar.

3. Las variables indexables empiezan desde 0, ya que el primer elemento es el que está "0 elementos lejos".

4. Las variables indexables son iterables, esto quiere decir que podemos pasar por cada uno de los valores gracias a los bucles, que veremos más adelante.

5. Se puede acceder al último elemento como variable[-1], al penúltimo como variable[-2], etc.

image.png

```python
x = int(input('Inserta el valor de la posición del único número escrito como cadena para poder imprimirlo: '))

lista_1 = ['adiós', 6, 25, "32"]

if lista_1[x].isnumeric():
  print(lista_1[x])
  print(type(lista_1[x]))
  print('¡Correcto!')
else:
  print('Revisa e inténtalo de nuevo')
```

Otra cosa que tienen en común las cadenas y las listas, es que se pueden "trocear" (slices) y sacar fragmentos de ellas, usando la misma idea de indexación

```python
palabra = "baldosa"

print( palabra[ 3:-1 ] )
```

lo que acaba de pasar es que imprimí un fragmento de la palabra baldosa, desde la cuarta letra (porque se empieza a contar desde 0) hasta la última (que nunca se cuenta, así que siempre queda hasta la letra anterior). También se puede usar en listas.

```python
listado = [1, 3, 5, 7, 9, 11]

print(listado[2:5])
```

### Funciones de listas

Existen varias funciones que nos permiten agregar, remover y manipularlas

#### Las esenciales

```python
lista = [1, 2, 3, 4, 5]
print(len(lista))
#len nos permite conocer la longitud de la lista
```

```python
lista.append(6) #agrega un 6 al final de la lista
print(lista, len(lista))
```

```python
lista.remove(1) #elimina el 1 de la lista (si está, de lo contrario, saldrá un error)
lista
```

```python
#también se puede remover por posición
lista.remove(lista[-2]) #elimina el penúltimo número
lista
```

```python
#pop permite eliminar un elemento de una lista y tenerlo guardado en una variable
#si no se coloca ninguna posición, elimina el último valor

numero = lista.pop(0)
numero, lista
```

```python
#se pueden insertar valores en posiciones específicas con insert
#primero se coloca la posición, y luego el valor a insertar

lista.insert(-1, 7) #insertar en la última posición de la lista un 7
lista
```

cuando se usa insert, se desplazan los demás valores a la derecha, es por eso que cuando se coloca en la "última posición", lo que hace es empujar el último valor, lo que lo deja en la penúltima oposición.

#### Otras funciones

```python
#max y min nos proporcionan los números mayores y menores de una lista

print(max(lista)) 
print(min(lista))
```

```python
print(sum(lista)) #podemos sumar todos los elementos de la lista
```

```python
#también clasifica en orden alfabético

otra_lista = ['3', '4', 'j', 'z', 'a', '1']
print(max(otra_lista), min(otra_lista))
```

```python
#organiza los elementos en orden ascendente

print(sorted(otra_lista))
```

# Operadores lógicos

Para sacar el mayor provecho a los condicionales, es importante saber usar estos operadores, ya que nos permiten combinar distintas posibilidades.

Por ejemplo, si quiero un programa que me diga si un número es par y positivo, podría hacer esto

```python
numero = int(input('Escribe un número: '))

if numero >= 0:
  if numero%2 == 0:
    print('El número es par y positivo')
```

Pero ese programa podría ser escrito en menos listas, y con menos anidaciones (siempre es preferible evitarlas, ya que hacen al código más complejo y pueden traer errores si están mal colocadas)

```python
numero = int(input('Escribe un número: '))

if numero > 0 and numero%2 == 0: #si el número es mayor a cero y no sobra nada cuando se divide entre 2
  print('El número es par y positivo')
```

image.png

Los operadores lógicos actúan como puertas que nos dan un resultado basado en un conjunto de condiciones, veremos las 4 más importantes.

Dato curioso: estas mismas puertas lógicas se usan en circuitos electrónicos y en computadores.

#### and 
este operador sirve cuando se quieren evaluar dos o más condiciones como verdaderas, si una de ellas es falsa, todo el argumento será falso

```python
if 3 > 1 and 3 > 2 and 3 >= 3 and 3/3 == 1:
  print('todo es correcto')
```

```python
if 4 > 1 and 4%2 == 0 and 4 < 2:
  print('todo correcto')
else:
  print('no todo es correcto')
```

#### or

este operador devuelve un verdadero si alguna o todas las condiciones son verdaderas

```python
if 3>4 or 4>3:
  print('al menos una condición es verdadera')
```

#### xor

es muy parecido a or, la única diferencia es que devuelve falso si todas las condiciones se cumplen (o no se cumplen), se escribe con este símbolo: ^

```python
if 3 < 4 ^ 4 > 3: 
  print('solo una condición es cierta')
else:
  print('o ambas condiciones son ciertas, o ninguna lo es')
```

#### not

representa el booleano contrario

```python
print(not True)
```

```python
if not 3 < 2: #si 3 no es menor a 2
  print('3 no es menor a 2')
```
